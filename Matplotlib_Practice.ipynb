# Matplotlib

```python
from matplotlib import pyplot as plt
import numpy as np
```

## Histogram

```python
blood_sugar_level_men = [83,130,97,88,114,112,135,90,111,132,105,93,97]
blood_sugar_level_women = [82,120,90,94,114,130,91,84,107,109,127,87,89]

plt.xlabel("Blood Sugar Level")
plt.ylabel("No. of Patients")
plt.title("Blood Sugar Analysis")
plt.hist([blood_sugar_level_men,blood_sugar_level_women],bins=[80,100,125,150],rwidth=0.95,color=['green','orange'],label=["men","women"])
plt.legend()
plt.show()
```

## Pie Chart

```python
expenses = [4000,2000,1500,1000,350]
categories = ['House Rent','Bills','Food & Beverages','Clothing','Others']
plt.pie(expenses,labels=categories,radius=1,explode=[0,0.1,0.1,0,0],shadow=True)
plt.show()
```

## Bar Graph

```python
xpos = np.arange(len(categories))
xpos
```

```python
plt.ylabel("Expenses in Rs")
plt.title("Household Finance")
plt.xticks(xpos,categories)
plt.bar(xpos,expenses) # can try many other attributes available here like color,width,label etc...
plt.show()
```
