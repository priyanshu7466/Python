```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/DeepLearningML/13_dropout_layer/sonar_dataset.csv", header= None)
df.head()
```

```python
X = df.drop(60, axis=1)
y = df[60]
```

```python
y = pd.get_dummies(df[60],drop_first=True)
y.sample(5)
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25,random_state=1)
```

```python
X_train.shape
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model = keras.Sequential([
    keras.layers.Dense(60,input_dim=60,activation='relu'),
    keras.layers.Dense(30,activation='relu'),
    keras.layers.Dense(15,activation='relu'),
    keras.layers.Dense(1,activation='sigmoid')
])

model.compile(
    optimizer = 'adam',
    loss = 'binary_crossentropy',
    metrics = ['accuracy']
)
model.fit(X_train,y_train,epochs=100,batch_size=8)
```

```python
model.evaluate(X_test,y_test)
```

```python
y_pred = model.predict(X_test).reshape(-1)
print(y_pred[:10])

y_pred = np.round(y_pred)
print(y_pred[:10])
```

```python
y_test[:10]
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
modeld = keras.Sequential([
    keras.layers.Dense(60,input_dim=60,activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(30,activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(15,activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(1,activation='sigmoid')
])

modeld.compile(
    optimizer = 'adam',
    loss = 'binary_crossentropy',
    metrics = ['accuracy']
)
modeld.fit(X_train,y_train,epochs=100,batch_size=8)
```

```python
modeld.evaluate(X_test,y_test)
```

```python
y_prediction = modeld.predict(X_test).reshape(-1)
print(y_prediction[:10])
y_prediction = np.round(y_prediction)
print(y_prediction[:10])
```

```python
y_test[:10]
```

```python
print(classification_report(y_test,y_prediction))
```
