```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/DeepLearningML/11_chrun_prediction/customer_churn.csv")
df.head()
```

```python
df.drop(['customerID'],axis='columns',inplace=True)
```

```python
df.head()
```

```python
for col in df:
  print(f' {col} : {df[col].unique()}')
```

```python
col_to_scale = ['tenure','MonthlyCharges','TotalCharges']
for col in df:
  print(f'{col} : {df[col].dtype}')
```

```python
df[df['TotalCharges']==" "].shape
```

```python
df.shape
```

```python
df1 = df[df['TotalCharges']!=" "]
df1.shape
```

```python
df1['TotalCharges'] = pd.to_numeric(df1['TotalCharges'])
```

```python
df1.dtypes
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df1[col_to_scale] = scaler.fit_transform(df1[col_to_scale])
```

```python
for col in df1:
  print(f'{col} : {df1[col].unique()}')
```

```python
df1.replace('No internet service','No',inplace=True)
```

```python
df1.replace('No phone service','No',inplace=True)
```

```python
df1.replace({'Yes': 1, 'No': 0}, inplace=True)
```

```python
col_to_encode = ['InternetService','Contract','PaymentMethod']
dummies = pd.get_dummies(df1[col_to_encode],drop_first=True)
df1 = pd.concat([df1,dummies],axis=1)
df1.head()
```

```python
df1.drop(col_to_encode ,axis=1 ,inplace = True)
df1.head()
```

```python
df1.replace({'Female' : 1 , 'Male' : 0}, inplace=True)
for col in df1:
  print(f'{col} : {df1[col].unique()}')
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
X = df1.drop('Churn',axis=1)
y = df1.Churn
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size=0.2,random_state=5)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
X_train.shape
```

```python
def ANN(X_train, X_test, y_train, y_test, loss, weights):
  model = keras.Sequential([
      keras.layers.Dense(23,input_dim=23,activation='relu'),
      keras.layers.Dense(15,activation='relu'),
      keras.layers.Dense(1,activation='sigmoid')
  ])
  model.compile(
      optimizer = 'adam',
      loss = loss,
      metrics = ['accuracy']
  )
  if weights==-1:
    model.fit(X_train,y_train,epochs=100)
  else:
    model.fit(X_train,y_train,epochs=100,class_weight = weights)

  print(model.evaluate(X_test,y_test))

  y_pred = model.predict(X_test)
  y_pred = np.round(y_pred)

  print("Classification Report \n", classification_report(y_test,y_pred))

  return y_pred
```

```python
y_pred = ANN(X_train,X_test,y_train,y_test,'binary_crossentropy',-1)
```

# Method 1: Undersampling

```python
class_count_0, class_count_1 = df1.Churn.value_counts()
print(class_count_0, class_count_1)

test_class_0 = df1[df1.Churn==0]
test_class_1 = df1[df1.Churn==1]
```

```python
test_class_under = test_class_0.sample(class_count_1)
test_class_under.shape
```

```python
test_class = pd.concat([test_class_under, test_class_1],axis=0)
print(test_class.Churn.value_counts())
```

```python
X = test_class.drop('Churn',axis=1)
y = test_class.Churn
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=15,stratify=y)
```

```python
y_train.value_counts()
```

```python
y_pred = ANN(X_train, X_test, y_train, y_test, 'binary_crossentropy',-1)
```

# Method2: Oversampling

```python
class_count_0, class_count_1 = df1.Churn.value_counts()
print(class_count_0, class_count_1)

test_class_0 = df1[df1.Churn==0]
test_class_1 = df1[df1.Churn==1]
```

```python
test_class_over = test_class_1.sample(class_count_0, replace=True)
test_class_over.shape
```

```python
test_class = pd.concat([test_class_0, test_class_over], axis=0)
test_class.shape
```

```python
X = test_class.drop('Churn',axis=1)
y = test_class.Churn
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=15,stratify=y)
```

```python
y_train.value_counts()
```

```python
y_pred = ANN(X_train, X_test, y_train, y_test, 'binary_crossentropy', -1)
```

# Method3: SMOTE

```python
X = df1.drop('Churn', axis=1)
y = df1.Churn
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(sampling_strategy='minority')
x_sm, y_sm = smote.fit_resample(X,y)

y_sm.value_counts()
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_sm,y_sm,test_size=0.2,random_state=15,stratify=y_sm)
```

```python
y_train.value_counts()
```

```python
y_pred = ANN(X_train, X_test, y_train, y_test, 'binary_crossentropy', -1)
```

# Method4: Use of Ensemble with undersampling

```python
X = df1.drop('Churn',axis=1)
y = df1.Churn
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=15,stratify=y)
```

```python
y_train.value_counts()
```

```python
df2 = X_train.copy()
df2['Churn'] = y_train
```

```python
df2.head()
```

```python
df2_class0 = df2[df2.Churn==0]
df2_class1 = df2[df2.Churn==1]
```

```python
def get_train_batch(df_majority, df_minority, start, end):
  df_train = pd.concat([df_majority[start:end], df_minority],axis=0)

  X_train = df_train.drop('Churn',axis=1)
  y_train = df_train.Churn

  return X_train, y_train
```

```python
X_train, y_train = get_train_batch(df2_class0, df2_class1, 0, 1495)
y_pred1 = ANN(X_train,X_test,y_train,y_test,'binary_crossentropy',-1)
```

```python
X_train, y_train = get_train_batch(df2_class0, df2_class1,1495,2990)
y_pred2 = ANN(X_train,X_test,y_train,y_test,'binary_crossentropy',-1)
```

```python
X_train, y_train = get_train_batch(df2_class0, df2_class1, 2990, 4130)
y_pred3 = ANN(X_train,X_test,y_train,y_test,'binary_crossentropy',-1)
```

```python
len(y_pred1)
```

```python
y_pred_final = y_pred1.copy()
for i in range(len(y_pred1)):
  if y_pred1[i]+y_pred2[i]+y_pred3[i] > 1:
    y_pred_final[i] = 1
  else:
    y_pred_final[i] = 0
```

```python
print(classification_report(y_test,y_pred_final))
```
