```python
from tvDatafeed import TvDatafeed,Interval
```

```python
tv = TvDatafeed()
```

```python
data_5_min = tv.get_hist(symbol='MARUTI',exchange='NSE',interval=Interval.in_5_minute,n_bars=5000)
df = pd.DataFrame(data_5_min)
print(df)
```

```python
nifty_data = tv.get_hist('NIFTY','NSE',interval=Interval.in_1_hour,n_bars=1000)
```
