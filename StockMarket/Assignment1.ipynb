```python
import numpy as np
import pandas as pd
from tvDatafeed import TvDatafeed, Interval
import talib as ta
```

```python
tv = TvDatafeed()
```

```python
data_5_min = tv.get_hist('MARUTI','NSE',interval=Interval.in_5_minute,n_bars=5000)
```

```python
df = pd.DataFrame(data_5_min)
df
```

```python
df['26 EMA'] = ta.EMA(data_5_min['close'],timeperiod=26)
df['200 EMA'] = ta.EMA(data_5_min['close'],timeperiod=200)
df['RSI'] = ta.RSI(data_5_min['close'])
df['datetime'] = df.index
df
```

```python
df = df.reset_index(drop=True)  # Drop the old index
df
```

```python
macd_series = ta.MACD(df.close,fastperiod=12,slowperiod=26,signalperiod=9)
macd, signal, hist = macd_series  # Unpack the tuple
df['MACD'] = pd.Series(signal)
df
```

```python
print(macd_series)
```

```python
df['Signal'] = np.where((macd > 0) & (df['close'] > df['200 EMA']), 'Buy',
                         np.where((macd < 0) & (df['close'] < df['200 EMA']), 'Sell', 'Hold'))
```
