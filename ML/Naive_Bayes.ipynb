# Titanic Survival Prediction

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/14_naive_bayes/titanic.csv")
df.head()
```

```python
df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns',inplace=True)
df.head()
```

```python
target = df.Survived
inputs = df.drop(['Survived'],axis='columns')
```

```python
inputs.head()
```

```python
dummies = pd.get_dummies(df.Sex)
dummies
```

```python
inputs = pd.concat([inputs,dummies],axis='columns')
inputs.head()
```

```python
inputs.drop(['Sex'],axis='columns',inplace=True)
inputs.head()
```

```python
inputs.Age.isna()
```

```python
inputs['Age'] = inputs.Age.fillna(inputs.Age.median())
inputs.head(10)
```

```python
inputs.Age.isna()
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model = GaussianNB()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inputs,target,test_size=0.25)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
X_test[:10]
```

```python
model.predict(X_test[:10])
```

```python
y_test[:10]
```

```python
model.predict_proba(X_test[:10])
```

```python
from sklearn.model_selection import cross_val_score
score = cross_val_score(GaussianNB(),X_train,y_train,cv=5)
np.average(score)
```

# Email Spam Filter

```python
df = pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/14_naive_bayes/spam.csv")
df.head()
```

```python
df.groupby(['Category']).describe()
```

```python
df['spam'] = df.Category.apply(lambda x: 1 if x=='spam' else 0)
df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.Message,df.spam)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
v = CountVectorizer()
```

```python
X_train_count = v.fit_transform(X_train.values)
```

```python
X_train_count.toarray()[:2]
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
```

```python
model.fit(X_train_count,y_train)
```

```python
emails = [
    'Hey mohan, can we get together to watch footbal game tomorrow?',
    'Upto 20% discount on parking, exclusive offer just for you. Dont miss this reward!'
]
emails_count = v.transform(emails)
model.predict(emails_count)
```

```python
X_test_count = v.transform(X_test)
model.score(X_test_count,y_test)
```

**sklearn pipeline**

```python
from sklearn.pipeline import Pipeline
```

```python
clf = Pipeline([
    ('vectorizer',CountVectorizer()),
    ('nb',MultinomialNB())
])
```

```python
clf.fit(X_train,y_train)
```

```python
clf.score(X_test,y_test)
```

```python
clf.predict(emails)
```
