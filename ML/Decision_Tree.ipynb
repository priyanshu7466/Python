```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/9_decision_tree/salaries.csv")
df.head()
```

```python
inputs = df[['company','job','degree']]
target = df['salary_more_then_100k']
```

```python
inputs
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
inputs['company_n'] = le.fit_transform(inputs.company)
inputs['job_n'] = le.fit_transform(inputs.job)
inputs['degree_n'] = le.fit_transform(inputs.degree)
inputs
```

```python
inputs_n = inputs.drop(['company','job','degree'],axis='columns')
inputs_n
```

```python
target
```

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
model.fit(inputs_n,target)
```

```python
model.score(inputs_n,target)
```

```python
model.predict([[2,2,1]])
```

```python
model.predict([[2,1,1]])
```
