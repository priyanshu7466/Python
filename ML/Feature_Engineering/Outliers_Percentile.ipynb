```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/FeatureEngineering/1_outliers/heights.csv")
df.head()
```

```python
df['height']
```

```python
max_threshold = df['height'].quantile(0.95)
max_threshold
```

```python
df[df['height']>max_threshold]
```

```python
min_threshold = df['height'].quantile(0.05)
min_threshold
```

```python
df[df['height']<min_threshold]
```

```python
df[(df['height']>min_threshold) & (df['height']<max_threshold)]
```
