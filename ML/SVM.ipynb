```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
dir(iris)
```

```python
iris.feature_names
```

```python
df = pd.DataFrame(iris.data,columns=iris.feature_names)
df.head()
```

```python
df['target'] = iris.target
df.head()
```

```python
df[df.target==2].head()
```

```python
df['flower name'] = df.target.apply(lambda x:iris.target_names[x])
df.head()
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df0 = df[df.target==0]
df1 = df[df.target==1]
df2 = df[df.target==2]
```

```python
plt.xlabel('sepal length (cm)')
plt.xlabel('sepal width (cm)')
plt.scatter(df0['sepal length (cm)'],df0['sepal width (cm)'],color='green',marker='+')
plt.scatter(df1['sepal length (cm)'],df1['sepal width (cm)'],color='blue',marker='.')
```

```python
plt.xlabel('petal length (cm)')
plt.xlabel('petal width (cm)')
plt.scatter(df0['petal length (cm)'],df0['petal width (cm)'],color='green',marker='+')
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'],color='blue',marker='.')
```

```python
from sklearn.svm import SVC
```

```python
model = SVC(C=10) # change attributes here like C , gamma , kernel
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(['target','flower name'],axis = 'columns')
X.head()
```

```python
y = df.target
```

```python
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size=0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
dir(digits)
```

```python
digits.target_names
```

```python
digits.target
```

```python
d = pd.DataFrame(digits.data,digits.target)
d.head()
```

```python
d['target'] = digits.target
d.head(20)
```

```python
X_train , X_test , y_train , y_test = train_test_split(d.drop(['target'],axis='columns'),d.target,test_size=0.2)
```

```python
len(X_train)
```

```python
Model = SVC(kernel='linear')
```

```python
Model.fit(X_train,y_train)
```

```python
Model.score(X_test,y_test)
```

```python
rbf_model = SVC(kernel='rbf')
```

```python
rbf_model.fit(X_train,y_train)
```

```python
rbf_model.score(X_test,y_test)
```
